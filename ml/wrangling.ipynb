```python
import pandas as pd
from pandarallel import pandarallel
pandarallel.initialize(nb_workers=24)
import dask.dataframe as dd
```

```python
clientsDF = pd.read_csv('./Data/Clientes_Descriptivo.txt', sep='\t')
clientsDF.head()
```

```python
clientsDFSynth = clientsDF.sample(frac=0.165, random_state=1337, weights='EDAD')
clientsDFSynth['acc'] = 1


clientsDFSynthDis = clientsDFSynth.sample(frac=0.388, random_state=1338, weights='EDAD')
clientsDFSynthDis['caminar'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)

clientsDFSynthDis = clientsDFSynth.sample(frac=0.61, random_state=1339, weights='EDAD')
clientsDFSynthDis['ver'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc', 'caminar'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)

clientsDFSynthDis = clientsDFSynth.sample(frac=0.244, random_state=1340, weights='EDAD')
clientsDFSynthDis['escuchar'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc', 'caminar', 'ver'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)

clientsDFSynthDis = clientsDFSynth.sample(frac=0.107, random_state=1341, weights='EDAD')
clientsDFSynthDis['hablar'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc', 'caminar', 'ver', 'escuchar'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)

clientsDFSynthDis = clientsDFSynth.sample(frac=0.116, random_state=1342, weights='EDAD')
clientsDFSynthDis['cuidado'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc', 'caminar', 'ver', 'escuchar', 'hablar'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)


clientsDFSynthDis = clientsDFSynth.sample(frac=0.237, random_state=1343, weights='EDAD')
clientsDFSynthDis['recordar'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc', 'caminar', 'ver', 'escuchar', 'hablar', 'cuidado'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)

clientsDFSynthDis = clientsDFSynth.sample(frac=0.076, random_state=1344, weights='EDAD')
clientsDFSynthDis['mental'] = 1
clientsDFSynthDis = clientsDFSynthDis.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO', 'acc', 'caminar', 'ver', 'escuchar', 'hablar', 'cuidado', 'recordar'])
clientsDFSynth = clientsDFSynth.join(clientsDFSynthDis)



clientsDFSynth = clientsDFSynth.drop(columns=['NU_CTE_COD', 'EDAD', 'CD_ESTADO', 'CD_POSTAL', 'CD_SEXO'])
clientsDF = clientsDF.join(clientsDFSynth)
clientsDF['acc'] = clientsDF['acc'].fillna(0)
clientsDF['mental'] = clientsDF['mental'].fillna(0)
clientsDF['recordar'] = clientsDF['recordar'].fillna(0)
clientsDF['cuidado'] = clientsDF['cuidado'].fillna(0)
clientsDF['hablar'] = clientsDF['hablar'].fillna(0)
clientsDF['escuchar'] = clientsDF['escuchar'].fillna(0)
clientsDF['ver'] = clientsDF['ver'].fillna(0)
clientsDF['caminar'] = clientsDF['caminar'].fillna(0)
clientsDF.describe()
```

```python
aggClientDF = pd.read_csv('./Data/clientsAgg.csv')
aggClientDF.head()
```

```python
aggClientDFPV = aggClientDF.pivot(index='NU_CTE_COD', columns='SubGiro', values=['mean', 'max', 'sum', 'count'])
aggClientDFPV.columns = ['_'.join(col) for col in aggClientDFPV.columns.values]
aggClientDFPV = aggClientDFPV.fillna(0)
aggClientDFPV.reset_index()
```

```python
clientsSynthAgg = clientsSynth.merge(aggClientDFPV, on='NU_CTE_COD')
clientsSynthAgg.to_csv('./Data/clientsSynthAgg.csv', index=False)
```

```python
aggClientMesDF = pd.read_csv('./Data/clientsAggMes.csv')
aggClientMesDF['Mes'] = aggClientMesDF['Mes'].astype(str)
aggClientMesDFPV = aggClientMesDF.pivot(index='NU_CTE_COD', columns=['SubGiro', 'Mes'], values=['mean', 'max', 'sum', 'count'])
aggClientMesDFPV.columns = ['_'.join(col) for col in aggClientMesDFPV.columns.values]
aggClientMesDFPV = aggClientMesDFPV.fillna(0)
aggClientMesDFPV.reset_index()
```

```python
clientsSynthAgg = clientsSynth.merge(aggClientMesDFPV, on='NU_CTE_COD')
clientsSynthAgg.to_csv('./Data/clientsSynthAgg.csv', index=False)
```
